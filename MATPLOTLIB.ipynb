```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x,y, 'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#00
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.4,0.4,0.4])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(2,2)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(1,2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(1,2)
axes[0].plot(y,x)
axes[0].set_title('First Plot')
plt.tight_layout()
```

```python
figg = plt.figure(figsize = (3,2))

ax = figg.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
figg = plt.figure(figsize = (8,2))

ax = figg.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
figg = plt.figure(figsize = (3,2))

ax = figg.add_axes([1,1,1,1])
ax.plot(x,y)
```

```python
figg,axes = plt.subplots(nrows =2, ncols =1, figsize = (8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
yem = plt.figure()

ax = yem.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Squared')
ax.plot(x,x**3, label = 'X Cubed')

ax.legend()
```

```python
zee = plt.figure()

ax = zee.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple', linewidth = 5, linestyle = '-.')
#linewidth can also be written as lw and alpha = 0.5 will fade the line
```

```python
zee = plt.figure()

ax = zee.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple', lw = 3, ls = ':')
#linewidth can also be written as lw and alpha = 0.5 will fade the line
```

```python
zee = plt.figure()

ax = zee.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple', linewidth = 1, linestyle = ':', marker = '*', markersize = 20)
#linewidth can also be written as lw and alpha = 0.5 will fade the line
```

```python
zee = plt.figure()

ax = zee.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple', linewidth = 1, linestyle = ':', marker = '*', markersize = 20,
       markerfacecolor = 'yellow', markeredgewidth =3, markeredgecolor = 'green')
#linewidth can also be written as lw and alpha = 0.5 will fade the line
```

```python
#PIE CHART
students = [40,55,100,80,25]
grade = ['A', 'B', 'C', 'D', 'E']
exp = [0,0,0.2,0,0]
plt.pie(students, labels = grade, explode = exp, autopct = '%2.1f%%')
plt.show()
```

```python
students = [40,55,100,80,25]
grade = ['A', 'B', 'C', 'D', 'E']
exp = [0,0,0.2,0,0]
plt.bar(grade,students, color = 'green', edgecolor = 'yellow')
plt.xlabel('grade')
plt.ylabel('students')
plt.title('Students Grade')
plt.show()
```
