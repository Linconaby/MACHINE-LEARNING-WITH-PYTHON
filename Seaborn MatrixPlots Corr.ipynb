```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
#these can be done using corr and pivot
```

```python
tc = tips.corr()
```

```python
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot = True)
```

```python
sns.heatmap(tc, annot = True, cmap = 'coolwarm')
```
